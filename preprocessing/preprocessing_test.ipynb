# MFCC
아예 다르게 생겼네.. 그래도 0.45 부분에서 특이하게 나오는 점들 이런건 다 보이네

```python
import librosa
import matplotlib.pyplot as plt
import os
import librosa.display

audio_dir = os.path.join('.','data','audio_label_clip')
y, sr = librosa.load(os.path.join(audio_dir,'u1722694_0.mp3'))

mfccs = librosa.feature.mfcc(y=y, sr=sr)
librosa.display.specshow(mfccs, x_axis='time')
```

```python
# spectrogram보다 frequency가 더 작음(1025->128)
mfccs.shape
```
